```python
from PIL import Image, ImageDraw
from tqdm import tqdm
import os

output_dir = r'C:\Users\AvivaLab\code\EEG_experiments\vMMR_EEG_PsychoPy_experiment\stimuli\faces'
os.makedirs(output_dir, exist_ok=True)

W, H = 425, 405
faces = {
    'self':           ('SELF',        180),
    'other':          ('OTHER',       140),
    'morph50':        ('MORPH 50%',   120),
    'self_target':    ('SELF\n[TGT]', 180),
    'other_target':   ('OTHER\n[TGT]',140),
    'morph50_target': ('MORPH\n[TGT]',120),
}

total = len(faces)
print(f'Starting generation of {total} images...')

for index, (filename, (label, gray_level)) in enumerate(
    tqdm(faces.items(), total=total, desc='Generating face stimuli', unit='image'),
    start=1,
):
    img = Image.new('L', (W, H), color=0)  # black background
    draw = ImageDraw.Draw(img)
    # Draw a grey oval as face placeholder
    draw.ellipse([50, 20, 375, 385], fill=gray_level)
    # Add label
    draw.text((W//2, H//2), label, fill=0, anchor='mm')
    img.save(os.path.join(output_dir, f'{filename}.png'))
    tqdm.write(f'[{index}/{total}] Saved {filename}.png')

print(f'Done. Wrote {total} files to {output_dir}')
```
