```python
import os

files = ['death_words.csv', 'negative_words.csv', 'neutral_words.csv']

for fname in files:
    # Read with original Windows-1255 encoding
    with open(fname, 'r', encoding='windows-1255') as f:
        content = f.read()
    
    # Strip any leading/trailing whitespace from each word
    words = [w.strip() for w in content.strip().splitlines() if w.strip()]
    
    # Write back as UTF-8 with header
    with open(fname, 'w', encoding='utf-8', newline='\n') as f:
        f.write('word\n')
        f.write('\n'.join(words))
        f.write('\n')
    
    print(f"{fname}: {len(words)} words, re-saved as UTF-8")
```
